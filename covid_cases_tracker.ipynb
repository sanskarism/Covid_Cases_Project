```python
import covid
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from covid import Covid
covid=Covid()
```

```python
name=input("Enter Country Name ")
virus_data=covid.get_status_by_country_name(name)

remove=['id','country','latitude','longitude','last_update']
for i in remove:
    virus_data.pop(i)
all = virus_data.pop('confirmed')
print(virus_data)

heading = list(virus_data.keys())
value= [str(i) for i in virus_data.values()]

# pie chart
plt.pie(value,labels=heading, colors = {'r','y','g','b'}, autopct='%1.1f%%')
plt.title("COUNTRY: "+name.upper() +"\n TOTAL CASES : "+str(all))
plt.legend()
plt.show()
```
